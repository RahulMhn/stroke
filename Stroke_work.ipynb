```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/Datasets/stroke.csv')
df
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df.drop(["id"],axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
df["bmi"]=df["bmi"].fillna(df["bmi"].mean())
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["gender"]=le.fit_transform(df["gender"])
df["ever_married"]=le.fit_transform(df["ever_married"])
df["work_type"]=le.fit_transform(df["work_type"])
df["Residence_type"]=le.fit_transform(df["Residence_type"])
df["smoking_status"]=le.fit_transform(df["smoking_status"])
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=4)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
y_pred_knn
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
print("Accuracy=",accuracy_score(y_test,y_pred_knn)*100)
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_knn))
print(classification_report(y_test,y_pred_knn))
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
```

```python
y_pred_svc=sv.predict(X_test)
y_pred_svc
```

```python
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
print("Accuracy=",accuracy_score(y_test,y_pred_svc)*100)
print(classification_report(y_test,y_pred_svc))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred_svc))
```

```python
import matplotlib.pyplot as plt
x=["knn","svc"]
y=[accuracy_score(y_test,y_pred_knn),accuracy_score(y_test,y_pred_svc)]
plt.bar(x,y)
plt.plot()
plt.show()
```
